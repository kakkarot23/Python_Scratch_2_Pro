### 48. Input number; print factorial.

```python
import math

num = int(input("Enter a non-negative integer: "))

if num < 0:
    print("Factorial is not defined for negative numbers.")
elif num == 0:
    print("The factorial of 0 is 1.")
else:
    factorial = math.factorial(num)
    print(f"The factorial of {num} is {factorial}.")
```

### 49. Input number of terms; print Fibonacci series.

```python
n_terms = int(input("Enter the number of terms for Fibonacci series: "))

a, b = 0, 1
count = 0

if n_terms <= 0:
    print("Please enter a positive integer.")
elif n_terms == 1:
    print("Fibonacci series up to 1 term:")
    print(a)
else:
    print("Fibonacci series:")
    while count < n_terms:
        print(a, end=" ")
        nth = a + b
        a = b
        b = nth
        count += 1
    print()
```

### 50. Input number; check if prime or not.

```python
num = int(input("Enter a number: "))

if num <= 1:
    print(f"{num} is not a prime number.")
elif num == 2:
    print(f"{num} is a prime number.")
else:
    is_prime = True
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(f"{num} is a prime number.")
    else:
        print(f"{num} is not a prime number.")
```

### 41. Input 3 angles; check if triangle is valid.

```python
angle1 = float(input("Enter first angle: "))
angle2 = float(input("Enter second angle: "))
angle3 = float(input("Enter third angle: "))
if angle1 > 0 and angle2 > 0 and angle3 > 0 and (angle1 + angle2 + angle3) == 180:
    print("The angles form a valid triangle.")
else:
    print("The angles do not form a valid triangle.")
```

### 42. Input cost price & selling price; print profit/loss.

```python
cost_price = float(input("Enter cost price: "))
selling_price = float(input("Enter selling price: "))

if selling_price > cost_price:
    profit = selling_price - cost_price
    print(f"Profit: {profit:.2f}")
elif selling_price < cost_price:
    loss = cost_price - selling_price
    print(f"Loss: {loss:.2f}")
else:
    print("No profit, no loss.")
```
