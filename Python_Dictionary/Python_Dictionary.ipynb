## Count Element Frequencies Using a Dictionary

### Subtask:
Generate Python code to count the frequency of elements in a list using a dictionary.

**Reasoning**:
The previous subtask of introducing how to count element frequencies has been completed. Now, I will generate Python code to demonstrate counting element frequencies in a list using a dictionary.



```python
my_list_to_count = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']

frequency_dict = {}
for item in my_list_to_count:
    frequency_dict[item] = frequency_dict.get(item, 0) + 1

print("Original List:", my_list_to_count)
print("Element Frequencies:", frequency_dict)
```
